```python
import pandas as pd
import numpy as np

#get data from site into XML
import requests
url = requests.get('https://en.wikipedia.org/wiki/List_of_largest_oil_and_gas_companies_by_revenue').text

# other libraries I'm interested in exploring
# https://www.dataquest.io/blog/web-scraping-tutorial-python/
from bs4 import BeautifulSoup
soup = BeautifulSoup(url,'lxml')
```

```python
soup
```

```python
ttable = soup.find('table',{'class':'wikitable sortable'})

table_headers = list()
for each_h in ttable.find_all('th'):
    table_headers.append(each_h.text[:-1])

countries = list()
ref_table = list()
for title in ttable.find_all('td'):
    span_t = title.find_all('span')
    ins_num = 1
    temp_flags = list()
    if len(span_t) > 0:
        for e_flag in span_t:
            temp_flags.append(e_flag.find('a').get('title'))
        
        countries.append('|'.join(temp_flags))
    else:
        ref_table.append(title)
print(len(countries))
```

```python
data = pd.DataFrame(columns=table_headers)
data['Country'] = countries
dis_run = 0
for i in np.arange(0,len(re_table),3):
    data['Company name'][dis_run] = re_table[i].text[:-1]
    data['Revenue 2017(US$ billion)'][dis_run] = ref_table[i+1].text[:-1]
    data['Revenue 2018(US$ billion)'][dis_run] = ref_table[i+2].text[:-1]
    dis_run+=1
data.head()
```
